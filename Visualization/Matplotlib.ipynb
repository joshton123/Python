# Matplotlib is a popular data visualization library in Python. It's often used for creating static, interactive, and animated visualizations in Pytho

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,200)
```

```python
y = np.sin(x)
```

```python
plt.plot(x,y)
```

```python
plt.xlabel('This is my x ddata')
plt.ylabel('This is my x ddata')
plt.title('this is title')
plt.plot(x,y)
plt.show()
```

```python
colours = np.random.rand(50)
sizes = 100 * np.random.rand(50)
x = np.random.rand(50)
y = np.random.rand(50)
plt.xlabel('x ddata')
plt.ylabel('y ddata')
plt.title('Random data')
plt.scatter(x,y,c=colours,s=sizes,alpha=.8)
```

```python
x = ['a','b','c','d','e']
y = np.random.rand(5)
plt.xlabel('x ddata')
plt.ylabel('y ddata')
plt.title('Random data')
plt.bar(x,y)
```

```python
x = ['a','b','c','d','e']
y = np.random.rand(5)
plt.xlabel('x ddata')
plt.ylabel('y ddata')
plt.title('Random data')
plt.barh(x,y)
```

```python
plt.figure(figsize=(6,2))
x = np.linspace(0,10,200)
y = np.sin(x)
plt.xlabel('x data')
plt.ylabel('y data')
plt.title('sin-wave data')
plt.plot(x,y,'--b')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(5,4))
colours = np.random.rand(50)
sizes = 100 * np.random.rand(50)
x = np.random.rand(50)
y = np.random.rand(50)
plt.xlabel('x ddata')
plt.ylabel('y ddata')
plt.title('Random data')
plt.scatter(x, y, c=colours, s=sizes, alpha=.8, marker='v')
plt.show()
```
